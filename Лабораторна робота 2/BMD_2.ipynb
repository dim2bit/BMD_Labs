```python
import matplotlib.pyplot as plt

%matplotlib inline
```

Створення вектора значень

```python
import numpy as np

x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

Побудова графіків

Функціональний метод

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y labal')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

Об'єктно-орієнтований метод

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])
axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Title')

axes2.plot(y,x,'b')
axes2.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Title')
```

```python
fig, axes = plt.subplots() 

axes.plot(x, y, 'r')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #одночасна побудова декількох графіків

axes[0].plot(x, y)
axes[1].plot(y, x)
```

```python
fig, axes = plt.subplots(nrows=4,ncols=4)
plt.tight_layout() #уникнення накладання графіків
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x ** 2, label="x^2")  # додавання легенди до графіку за допомогою label
ax.plot(x, x ** 3, label="x^3")

ax.legend()
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)  # зміна розміру та DPI
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x ** 2, label="x^2")
ax.plot(x, x ** 3, label="x^3")

ax.legend()
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, lw=2, ls='--', color='b') # ширина, тип та колір лінії
ax.plot(y, x, lw=3, color='r')
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, marker='o', markersize=10, markerfacecolor='y') # тип, розмір та колір маркерів
```

```python
fig.savefig("figure.png")  #збереження графіка
```
