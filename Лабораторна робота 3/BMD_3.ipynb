```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])  # розподіл одновимірного набору спостережень
```

```python
sns.distplot(tips['total_bill'], bins = 50) # ширина кроку для гістограми
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips)  # діаграма розсіювання для визначення взаємозвязку між 2 змінними
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips,kind='kde')  # kind - варіант діаграми розсіювання
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips,kind='hex')
```

```python
sns.pairplot(tips)  # попарні взаємозвязки між змінними в повному наборі даних
```

```python
sns.pairplot(tips,hue='sex')  # hue - вибір категоріальної змінної для розділення даних на окремі підгрупи
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)  # кількісні дані в залежності від категоріальної змінної
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.median,hue='sex') # зміна варіанту відображення кількісної змінної
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow') # розподіл категоріальних даних за певною кількісною змінною
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')  # діаграма розсіювання, де одна змінна категоріальна
```

```python
sns.heatmap(tips.corr())  # кореляція між даними у вигляді матриці
```

```python
import matplotlib.pyplot as plt

g = sns.FacetGrid(tips,col="time",row='smoker')  # площа відображення даних для декільнох кількісних змінних
g = g.map(plt.hist,'total_bill')
```

```python
g = sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
g = g.map(plt.hist,'total_bill').add_legend()
```
